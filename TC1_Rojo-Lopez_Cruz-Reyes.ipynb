###### La funcion buscaMosqueteros(tablero) se encarga de encontrar y devolver las ubicaciones de los 3 msoqueteros en el tablero

```python
def buscaMosqueteros(tablero):
    ubicaciones = []
    for i in range(0,len(tablero)):
        for j in range(0,len(tablero[0])):
            if tablero[i][j] == 10:
                cor = [i,j]
                ubicaciones.append(cor)
    return ubicaciones
```

###### La funcion buscaEspacioEnBlanco como su nombre lo indica busca los espacion vacios del tablero y la funcion guardiasAlrededor verifica si alrededor de una casilla existen guardias

```python
def buscaEspacioEnBlanco(tablero):
    blancos = []
    for i in range(0,len(tablero)):
        for j in range(0,len(tablero[0])):
            if tablero[i][j] == 0:
                cor = [i,j]
                blancos.append(cor)
    return blancos

def guardiasAlrededor(x,y,tablero):
    guardias = []
    pos_validas = valida_coordenadas(x,y)
    for i in pos_validas:
        if tablero[i[0]][i[1]] == 1:
            guardias.append(i)
    return guardias
```

###### La funcion generaJugadasMosqueteros(tablero) se encarga de obtener todas las posibles jugadas de los 3 mosqueteros en el tablero

```python
import copy
def generaJugadasMosqueteros(tablero):
    jugadas = []
    tablero_copia = []
    posiciones_mosqueteros = buscaMosqueteros(tablero)
    jugadas_mos1 = obtenerPosiblesMovimientosMosquetero(posiciones_mosqueteros[0][0],posiciones_mosqueteros[0][1],tablero)
    jugadas_mos2 = obtenerPosiblesMovimientosMosquetero(posiciones_mosqueteros[1][0],posiciones_mosqueteros[1][1],tablero)
    jugadas_mos3 = obtenerPosiblesMovimientosMosquetero(posiciones_mosqueteros[2][0],posiciones_mosqueteros[2][1],tablero)
    
    tablero_copia = copy.deepcopy(tablero)
    
    if len(jugadas_mos1)!=0:
        for i in jugadas_mos1:
            tablero_futuro = copy.deepcopy(tablero_copia)
            tablero_futuro[i[0]][i[1]] = 10
            tablero_futuro[posiciones_mosqueteros[0][0]][posiciones_mosqueteros[0][1]] = 0
            jugadas.append(tablero_futuro)
            
    if len(jugadas_mos2)!=0:
        for i in jugadas_mos2:
            tablero_futuro = copy.deepcopy(tablero_copia)
            tablero_futuro[i[0]][i[1]] = 10
            tablero_futuro[posiciones_mosqueteros[1][0]][posiciones_mosqueteros[1][1]] = 0
            jugadas.append(tablero_futuro)
            
    if len(jugadas_mos3)!=0:
        for i in jugadas_mos3:
            tablero_futuro = copy.deepcopy(tablero_copia)
            tablero_futuro[i[0]][i[1]] = 10
            tablero_futuro[posiciones_mosqueteros[2][0]][posiciones_mosqueteros[2][1]] = 0
            jugadas.append(tablero_futuro)      
    return jugadas
```

###### La funcion generaJugadasGuardia(tablero)  se encarga de obtener todas las posibles jugadas de los guardias que quedene en pie en el tablero

```python
import copy
def generaJugadasGuardia(tablero):
    jugadas = []
    tablero_copia=[]
    lista = []
    guardias = []
    blancos = buscaEspacioEnBlanco(tablero)
    tablero_copia = copy.deepcopy(tablero)
    for i in blancos:
        lista = guardiasAlrededor(i[0],i[1],tablero)
        for i in lista:
            guardias.append(i)  
    
    for i in guardias:
        jugada_guardia = obtenerPosiblesMovimientosGuardia(i[0],i[1],tablero)
        for mov in jugada_guardia:
            tablero_futuro = copy.deepcopy(tablero_copia)
            tablero_futuro[mov[0]][mov[1]] = 1
            tablero_futuro[i[0]][i[1]] = 0
            jugadas.append(tablero_futuro)
    return jugadas
```

###### La funcion funcionEvaluacionMosqueteros(tablero)  se encarga de darle un valor al tablero dado. Esta funcion es la que se usara para evaluar que jugada le conviene mas al mosquetero para ganarle al guardia obteneindo la distancia entre ellos y sumandolas

```python
import math
def funcionEvaluacionMosqueteros(tablero):
    ubicaciones = buscaMosqueteros(tablero)
    distancia1 = math.sqrt((ubicaciones[0][0] - ubicaciones[1][0])**2 + (ubicaciones[0][1] - ubicaciones[1][1])**2)
    distancia2 = math.sqrt((ubicaciones[1][0] - ubicaciones[2][0])**2 + (ubicaciones[1][1] - ubicaciones[2][1])**2)
    distancia3 = math.sqrt((ubicaciones[0][0] - ubicaciones[2][0])**2 + (ubicaciones[0][1] - ubicaciones[2][1])**2)
    suma = distancia1 + distancia2 + distancia3
    return suma
```

###### La funcion victoria(tablero) permite determinar el ganador de la partida evaluando el tablero despues de cada jugada realizada ya sea por el mosquetero o la guardia

```python
def victoria(tablero):
    ubicaciones = []
    cantidad_guardias = 0
    ganador = False
    
    verifica_fila = False
    verifica_columna = False
    guardiasVivos = True
    sinMovimientos = False
    
    # mapeo del estado actual del tablero
    for i in range(0,len(tablero)):
        for j in range(0,len(tablero[0])):
            if tablero[i][j] == 10:
                cor = [i,j]
                ubicaciones.append(cor)
            if tablero[i][j] == 1:
                cantidad_guardias = cantidad_guardias + 1
                
    # verifica si todos los guardias fueron eliminados
    if cantidad_guardias == 0:
        guardiasVivos = False
    
    mosquetero1 = obtenerPosiblesMovimientosMosquetero(ubicaciones[0][0],ubicaciones[0][1],tablero)
    mosquetero2 = obtenerPosiblesMovimientosMosquetero(ubicaciones[1][0],ubicaciones[1][1],tablero)
    mosquetero3 = obtenerPosiblesMovimientosMosquetero(ubicaciones[2][0],ubicaciones[2][1],tablero)
    
    # verifica si los mosqueteros estan en la misma fila
    if ubicaciones[0][0] == ubicaciones[1][0] == ubicaciones[2][0]:
        verifica_fila = True
        # verifica si los mosqueteros estan en la misma columna
    if ubicaciones[0][1] == ubicaciones[1][1] == ubicaciones[2][1]:
        verifica_columna = True
    
    
    if (len(mosquetero1) == 0 ) and (len(mosquetero2) == 0 ) and (len(mosquetero3) == 0 ):
        sinMovimientos = True
    
    # COMPROBAMOS SI LOS GUARDIAS GANAN
    if (verifica_fila == True) or (verifica_columna == True):
        print("--------------------------GANAN LOS GUARDIAS------------------------------")
        return True
    else:
        if (guardiasVivos == False) or (sinMovimientos == True):
            print("--------------------------GANAN LOS MOSQUETEROS------------------------------")
            return True
        else:
            print("--------------------------CONTINUA EL JUEGO------------------------------")
            return False
```
